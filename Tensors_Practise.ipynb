```python
import tensorflow as tf
import numpy as np
```

# Create a tensor of shape (3, 3) with random numbers between 0 and 1

```python
tf.random.set_seed(2)
t1 = tf.random.uniform(shape=(3, 3), minval=0, maxval=1)
print(f"First Tensor Created")
print(t1)
```

# Perform basic tensor operations including matrix addition, multiplication, and transpose

```python
tf.random.set_seed(3)
t2 = tf.random.uniform(shape=(3, 3), minval=0, maxval=1)
print(f"Second Tensor Created")
print(t2)
```

```python
# Perform matrix addition
addition = tf.add(t1, t2)
print(f"Matrix Addition")
addition
```

```python
# Perform matrix multiplication
multiplication = tf.matmul(t1, t2)
print(f"Matrix multiplication")
print(multiplication)
```

```python
# Transpose
transpose = tf.transpose(t1)
print(f"Transpose of First Tensor")
print(transpose)
transpose2 = tf.transpose(t2)
print(f"\nTranspose of Seoncd Tensor")
print(transpose2)
```

# Convert the resulting tensor into a NumPy array and then reshape it into a (1, 9) tensor.

```python
# Convert tensor to NumPy array
numpy = t1.numpy()
print(f"Numpy Created")
print(numpy)
```

```python
# Reshape the NumPy array to (1, 9)
reshaped = numpy.reshape(1, 9)
print(f"Numpy reshaped")
print(reshaped)
```
