```python
import nltk
from nltk import word_tokenize
```

NLTK - Natural Language Toolkit

* is an open-source library. it provides many functionalities and include pre- trained models

* integrates well with other libraries, such as skit-learn

```python
nltk.download('punkt') # it is pretrained model for tokenization
```

```python
nltk.download('punkt_tab')
```

Perform Tokenization

```python
sentence = '"Today is friday. Next Week , We’re coming from U.S.!"'
```

```python
for token in word_tokenize(sentence):
  print(token)
```

```python
sent2 = "we're  here to help you! ask the question, email support@abcdnet.ax or visit us at http://www.abc.net.ax! "
```

```python
for token in word_tokenize(sent2):
  print(token)
```

```python
sent3 = "we are learning NLP, this is a good course. it weather good today?"
```

```python
for token in word_tokenize(sent3):
  print(token)
```
